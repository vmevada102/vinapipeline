```python
# =========================================================
# AUTODOCK VINA LIGAND PREPARATION PIPELINE
# GOOGLE COLAB ONE-CELL COMPLETE WORKFLOW
# =========================================================

# STEP 1 — INSTALL CONDACOLAB
!pip install -q condacolab

# STEP 2 — INSTALL CONDA
import condacolab
condacolab.install()

# =========================================================
# IMPORTANT
# =========================================================
# AFTER THIS CELL RUNS:
#
# Google Colab WILL RESTART automatically.
#
# AFTER RESTART:
# RUN THIS SAME CELL AGAIN.
#
# On second run:
# - conda already exists
# - dependencies install
# - script downloads
# - ligand upload starts
# - pipeline runs
# =========================================================

import os
import subprocess
import shutil

# =========================================================
# INSTALL DEPENDENCIES
# =========================================================

print("\nInstalling dependencies...\n")

subprocess.run(
    "conda install -c conda-forge rdkit pandas gemmi scipy -y",
    shell=True,
    check=True
)

subprocess.run(
    "pip install meeko",
    shell=True,
    check=True
)

subprocess.run(
    'pip install "numpy<2"',
    shell=True,
    check=True
)

print("\nDependencies installed successfully.\n")

# =========================================================
# DOWNLOAD SCRIPT
# =========================================================

print("\nDownloading ligand preparation script...\n")

subprocess.run(
    "wget -O colab_convert.py https://raw.githubusercontent.com/vmevada102/vinapipeline/main/colab_convert.py",
    shell=True,
    check=True
)

print("\nScript downloaded successfully.\n")

# =========================================================
# CREATE INPUT FOLDER
# =========================================================

os.makedirs("input", exist_ok=True)

# =========================================================
# UPLOAD FILES
# =========================================================

from google.colab import files

print("\nUpload ligand files now.\n")

uploaded = files.upload()

# =========================================================
# MOVE FILES
# =========================================================

for f in uploaded.keys():

    shutil.move(
        f,
        os.path.join("input", f)
    )

print("\nFiles moved to input/ successfully.\n")

# =========================================================
# RUN PIPELINE
# =========================================================

print("\nStarting ligand preparation pipeline...\n")

# Modified to capture stdout and stderr for better debugging and provide 'yes' input
result = subprocess.run(
    "echo 'yes' | python colab_convert.py", # Pipe 'yes' to stdin
    shell=True,
    check=False, # Set check=False so we can inspect output even if it fails
    capture_output=True,
    text=True
)

if result.returncode != 0:
    print(f"Error running colab_convert.py (Exit Code: {result.returncode})")
    print("--- Standard Output ---")
    print(result.stdout)
    print("--- Standard Error ---")
    print(result.stderr)
    # Re-raise the exception to stop execution and indicate failure
    raise subprocess.CalledProcessError(result.returncode, result.args, output=result.stdout, stderr=result.stderr)
else:
    print("\nPipeline completed successfully.\n")
    print("--- Script Output ---")
    print(result.stdout)


# =========================================================
# ZIP OUTPUT
# =========================================================

print("\nCreating ZIP archive...\n")

subprocess.run(
    "zip -r ligands.zip ligands",
    shell=True,
    check=True
)

print("\nZIP archive created.\n")

# =========================================================
# DOWNLOAD RESULTS
# =========================================================

print("\nDownloading results...\n")

files.download("ligands.zip")

# Optional downloads
if os.path.exists("ligand_properties.csv"):

    files.download(
        "ligand_properties.csv"
    )

if os.path.exists("error_log.txt"):

    files.download(
        "error_log.txt"
    )

if os.path.exists("error_convert.txt"):

    files.download(
        "error_convert.txt"
    )

print("\nALL TASKS COMPLETED SUCCESSFULLY.\n")
```
